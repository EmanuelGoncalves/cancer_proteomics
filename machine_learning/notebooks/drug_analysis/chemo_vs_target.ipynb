```python
import pandas as pd
```

```python
drug_info = pd.read_csv("../../data/drug/archive/screened_compunds_rel_8.2.csv")
drug_info['DRUG_ID'] = drug_info['DRUG_ID'].astype(str)
drug_type_map = drug_info.set_index('DRUG_ID').to_dict()['Drug_Type']
```

```python
drug_scores = pd.read_csv("../../data/drug/dl_ttest_df.csv")
```

```python
drug_scores['Drug_Type'] = drug_scores['drug_id'].map(lambda x: drug_type_map[
    x.split(";")[0]] if x.split(";")[0] in drug_type_map else None)
```

```python
drug_scores = drug_scores[~pd.isna(drug_scores['Drug_Type'])]
```
